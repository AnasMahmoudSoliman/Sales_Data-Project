```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\anas.mahmoud\\Downloads\\sales_data.csv');
```

```python
print (df)
```

```python
df['revenue'] = pd.to_numeric(df['revenue'])
total_revenue = df['revenue'].sum()
print(f"the total revenue is : ${total_revenue:.2f}")
```

```python
product_revenue = df.groupby('product')['revenue'].sum()
max_revenue_product = product_revenue.idxmax()
max_revenue = product_revenue.max()
plt.figure(figsize=(8, 4))
plt.bar(product_revenue.index, product_revenue, color='skyblue')
plt.xlabel('Product')
plt.ylabel('Revenue ($)')
plt.title(f'Product with Highest Revenue: {max_revenue_product} (${max_revenue:.2f})')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['price'] = pd.to_numeric(df['price'])
average_price = df['price'].mean()
print(f"The average price of a product sold by the company is ${average_price:.2f}")
```

```python
df['quantity'] = pd.to_numeric(df['quantity'])
total_quantity = df['quantity'].sum()
print(f"the total quantity = {total_quantity}")
```

```python
df['revenue'] = pd.to_numeric(df['revenue'])
category_revenue = df.groupby('category')['revenue'].sum()
max_revenue_category = category_revenue.idxmax()
max_revenue = category_revenue.max()
plt.figure(figsize=(8, 4))
plt.bar(category_revenue.index, category_revenue, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Revenue ($)')
plt.title(f'Category with Highest Revenue: {max_revenue_category} (${max_revenue:.2f})')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
total_sales = len(df)
average_sale_revenue=total_revenue / total_sales
print(f"The average sale rvenue = ${average_sale_revenue:.2f}")
```

```python
df['date'] = pd.to_datetime(df['date'])
df['quarter'] = df['date'].dt.to_period('Q')
quarterly_revenue = df.groupby('quarter')['revenue'].sum()
print("Total revenue generated in each quarter:")
print (quarterly_revenue)
plt.figure(figsize=(8, 4))
quarterly_revenue.plot(kind='bar', color='skyblue')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue Generated in Each Quarter')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
